### 2. series 클래스와 dataframe 클래스에 대해 설명

##### 시리즈 클래스
> 1차원의 시계열 데이터를 표현하고자 할 때 사용하는 클래스로 인덱스와 값이 한 쌍으로 나열되어있는 형태
> 시리즈를 생성하는 방법 - pandas 패키지의 `Series` 클래스의 생성자로 값과 인덱스에 대한 배열 혹은 리스트를 전달하면 생성할 수 있음 
> 시리즈도 numpy 배열과 같이 벡터화 연산이 가능함.(단, 연산 작업은 값에만 적용됨)

##### 데이터프레임 클래스
> 2차원 형태의 행렬에 열과 행에 대한 인덱스가 붙은 형태. 즉, 표(table)
1. 열 인덱스 배열 혹은 리스트를 생성
2. 행 인덱스 배열 혹은 리스트를 생성
3. 각 열과 행에 해당하는 데이터를 가지고 있는 딕셔너리를 생성
4. pandas 패키지의 DataFrame 클래스의 생성자로 데이터 딕셔너리와 행 인덱스, 열 인덱스를 전달하여 생성

### 3. numpy 배열에서 벡터화 연산이 파이썬 기본 리스트를 사용한 연산보다 어떤점에서 장점인지 설명

numpy 배열의 벡터화 연산은 모든 종류의 연산에 적용이 가능
numpy 배열은 반복문을 쓰지않고 배열에 대하여 벡터 연산을 할 수 있음.

### 4. 라인 플롯, 바 플롯, 스캐터 플롯, 파이플롯에 대해 간단히 설명하고 어떤 경우에 사용되는지 설명

##### 라인 플롯
> 시간의 흐름에 따른 데이터 포인터를 선으로 연결하여 나타내는 방식
> 시간의 흐름에 따라 비교할 때 주로 사용됨 (기온의 변화, 주식 가격)

##### 바 플롯
> 데이터의 카테고리별 값을 막대로 나타내는 방식 (수직, 수평)  
> 카테고비렬로 비교할 때 주로 사용됨 (제품별 판매량, 연도별 인구수)

##### 스캐터 플롯
> 두 변수 간의 상관 관계를 점들로 분포로 나타내는 방식
> 두 변수의 상관관계를 파악할 때 주로 사용됨 (키와 몸무게의 상관관계)

##### 파이 플롯
> 전체에 대한 비율을 원형으로 나타내는 방식 (각 조각의 크기가 해당 부분의 비율)
> 전체에 대한 비율을 비교할 때 주로 사용됨 (여론조사 결과)

### 6. numpy 패키지와 pandas 패키지에 대해 설명

#### 넘파이 패키지
> 파이썬 리스트의 단점인 원소의 자료형 지정 불가, 많은 메모리 차지의 문제를 해결하고자 수치 해석 프로그램에서는 numpy라고 하는 배열 패키지를 사용함.
> numpy 배열의 경우 자체적 C언어로 구현되어 있어 자료형 지정과 메모리 절약 효과를 볼 수 있음.

#### 판다스 패키지
> 데이터를 시계열이나 표로 표현하기 위한 패키지  
> 시계열을 표현하는 `Series` 클래스와 표로 표현하는 `DataFrame` 클래스가 존재함
