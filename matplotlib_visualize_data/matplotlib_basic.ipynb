### 데이터 시각화 패키지 맷플롯리브
matplotlib 패키지는 파이썬에서 데이터를 차트 혹은 플롯으로 시각화하는 패키지
- 선형 그래프 (line plot)
- 점 그래프 (scatter plot)
- 등고선 그래프 (contour plot)
- 표면도 (surface plot)
- 막대 그래프 (bar chart)
- 히스토그램 (histogram)
- 박스 그래프(box plot)

### plplot 하위 패키지

### 차트의 종류

**막대 그래프** : 데이터의 카테고리별 값을 막대로 나타내는 방식 (수직, 수평)  
> 카테고비렬로 비교할 때 주로 사용됨 (제품별 판매량, 연도별 인구수)  

**선 그래프** : 시간의 흐름에 따른 데이터 포인터를 선으로 연결하여 나타내는 방식
> 시간의 흐름에 따라 비교할 때 주로 사용됨 (기온의 변화, 주식 가격)  

**파이 차트** : 전체에 대한 비율을 원형으로 나타내는 방식 (각 조각의 크기가 해당 부분의 비율)
> 전체에 대한 비율을 비교할 때 주로 사용됨 (여론조사 결과)

**산점도** : 두 변수 간의 상관 관계를 점들로 분포로 나타내는 방식
> 두 변수의 상관관계를 파악할 때 주로 사용됨 (키와 몸무게의 상관관계)

**히스토그램** : 범위별로 그룹화해서 각 그룹에 속하는 데이터 포인터의 수를 막대로 나타내는 방식
> 데이터의 분포를 비교할 때 주로 사용됨 (시험 점수의 분포)

**영역 그래프** : 선 그래프의 아래 영역에 색으로 채운 형태  
> 시간의 흐름에 따라 비교할 때 주로 사용됨, 변화량을 강조할 때 주로 사용됨 (누적 감염자,누적 수출량)

**박스 그래프** : 데이터의 분포를 나타내는 방식, 최소값, 최대값, 사분위을 나타내는 방식
> 데이터의 이상치를 판단할 때 주로 사용됨

### 틱 설정
`xticks()`, `yticks()` 메서드로 틱의 위치나 라벨을 직접 지정할 수 있음. 

````python
X = np.linspace(-np.pi,np.pi,256)
C = np.cos(X)

plt.title('틱 설정')
plt.plot(X,C)
plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi])
plt.yticks([-1, 0, +1],['Low', 'Zero', 'High'])
plt.show()
````

### Grid 설정
`Grid()` 메서드로 그리드를 표현할지 말지를 정할 수 있음

````python
X = np.linspace(-np.pi,np.pi,256)
C = np.cos(X)

plt.title('틱 설정')
plt.plot(X,C)
plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi])
plt.yticks([-1, 0, +1],['Low', 'Zero', 'High'])
plt.grid(True)
plt.show()
````

### 여러 그래프를 하나의 차트에 표현하기
`plot()` 메서드에 x, y, style 매개변수를 반복해서 전달하여 여러 그래프를 하나의 차트에 표현할 수 있음 (스타일 문자열 생략 불가능)

````python
t = np.arange(0., 5., 0.2)
plt.title('여러 그래프 표현')
plt.plot(t, t, 'r--', t, 0.5 * t**2, 'bs:', t, 0.2 * t**3, 'g^-')
plt.grid(True)
plt.show()
````

### 범례 
여러 개의 그래프가 한 차트에 표시될 때 각 그래프가 무엇을 의미하는지 보여주는 범례를 표현할 수 있음
`legend()` 메서드로 표현 (`plot()`의 `label` 매개변수를 지정해야함), 위치를 수동으로 지정하고 싶으면 `loc` 매개변수를 사용함.

````python
plt.title('범례 표현')
plt.plot([2000,2005,2010,2015,2020,2025],[1, 2, 4, 8, 16, 32], lw=5, c='#00998850', ls=':', marker='*', ms=10, mfc='#009988', label='증가')
plt.plot([2000,2005,2010,2015,2020,2025],[32, 16, 8, 4, 2, 1], lw=5, c='#0099', ls=':', marker='o', ms=10, mfc='#008400', label='감소')
plt.legend(loc=10)
plt.show()
````

### 차트 그림의 구조
matplotlib으로 그려지는 차트는 Figure 객체, Axes 객체, Axis 객체로 구성이 되어져있음.  
Figure 객체는 하나 이상의 Axes 객체를 포함하고 있고, Axes 객체는 두개 이상의 Axis 객체를 포함하고 있음.  
- `Figure` : 차트가 그려지는 캔버스
- `Axes` : 하나의 plot
- `Axis` : 하나의 축

#### Figure 객체
모든 차트 그래프 결과는 Figure 객체에 포함되어 있음  
내부 차트가 아닌 경우 하나의 Figure 객체에는 하나의 아이디와 윈도우를 가짐  
일반적으로 기본 파이썬에서 차트를 표시하는 Figure 하나당 하나의 윈도우를 가짐  

`plot()` 메서드가 자동으로 Figure 객체를 생성하기 때문에 직접 생성하는 경우는 드물지만  
여러 개의 윈도우를 동시에 출력할 때나 차트 전체의 크기를 변경하고 싶을 때 사용할 수 있음

````python
plt.figure(figsize=(10, 2))
plt.title('figure 객체 사이즈 변경')
plt.plot(np.random.randn(100))
plt.show()
````

현재 사용하고 있는 Figure 객체를 얻으려면 `gcf()` 메서드로 얻을 수 있음

````python
f1 = plt.figure(figsize=(10, 2))
plt.title('figure 객체 사이즈 변경')
plt.plot(np.random.randn(100))

f2 = plt.gcf()
print(id(f1))
print(id(f2))
plt.show()

plt.title('figure 객체 사이즈 변경')
plt.plot(np.random.randn(100))

f3 = plt.gcf()
print(id(f3))
````

### `Axes` 객체와 subplot
하나의 figure 객체 안에 여러 개의 Axes 객체를 생성하여 행렬 형태의 subplot을 생성할 수 있음  
이때 `subplot` 메서드를 사용하여 명시적으로 Axes 객체를 생성할 수 있음  

그러나, pyplot의 `plot()` 메서드를 바로 사용해도 Axes의 객체를 자동으로 생성할 수 있음

````python
x1 = np.linspace(0,5)
x2 = np.linspace(0,2)
y1 = np.cos(2*np.pi*x1)*np.exp(-x1)
y2 = np.cos(2*np.pi*x2)

plt.subplot(2,1,1)
plt.plot(x1,y1,'yo-')
plt.title('2행 1열 서브 플롯 첫 번째')

plt.subplot(2,1,2)
plt.plot(x2,y2,'r.-')
plt.title('2행 1열 서브 플롯 두 번째')

plt.tight_layout()
plt.show()
````

`subplot()` 메서드의 (x,y,index)의 인수를 (xyindex)로 붙여서 사용할 수 있음  
ex) `subplot(2,2,1)` -> `subplot(221)`

````python
plt.subplot(221)
plt.plot(np.random.rand(5))
plt.title('첫 번째 플롯')

plt.subplot(222)
plt.plot(np.random.rand(5))
plt.title('두 번째 플롯')

plt.subplot(223)
plt.plot(np.random.rand(5))
plt.title('세 번째 플롯')

plt.subplot(224)
plt.plot(np.random.rand(5))
plt.title('네 번째 플롯')

plt.tight_layout()
plt.show()
````

`subplots(x,y)` 메서드로 복수의 Axes 객체를 담은 배열을 생성할 수 있음
`subplot(x,y)` 메서드는 (figure, axes 배열) 튜플을 반환

````python
figure, axes = plt.subplots(2,2)

axes[0,0].plot(np.random.rand(5))
axes[0,0].set_title('1행 1열')

axes[0,1].plot(np.random.rand(5))
axes[0,1].set_title('1행 2열')

axes[1,0].plot(np.random.rand(5))
axes[1,0].set_title('2행 1열')

axes[1,1].plot(np.random.rand(5))
axes[1,1].set_title('2행 2열')

plt.tight_layout()
plt.show()
````

#### Axes 객체와 축
`twinx()` 메서드로 x축을 공유하는 새로운 Axes 객체를 생성할 수 있음

````python
figure, axes0 = plt.subplots()
axes1 = axes0.twinx()

axes0.set_title('x축을 공유하는 플롯')

axes0.plot([10,5,2,9,7], 'r-', label="y0")
axes0.set_ylabel('y0')

axes1.plot([100,200,220,180,120],'g:',label='y1')
axes1.set_ylabel('y1')

axes0.set_xlabel('x축')

plt.show()
````
